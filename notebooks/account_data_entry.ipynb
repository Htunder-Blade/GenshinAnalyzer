```python
from __future__ import annotations

from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

SRC_PATH = PROJECT_ROOT / "src"
if str(SRC_PATH) not in sys.path:
    sys.path.insert(0, str(SRC_PATH))

from sqlalchemy import or_, select

from genshin_analyser.db import init_account_db, init_db, make_account_session_factory, make_session_factory
from genshin_analyser.models import Account, AccountArtifact, AccountCharacter, AccountWeapon, ArtifactSet
from genshin_analyser.repository import (
    count_account_artifacts,
    count_account_characters,
    count_account_weapons,
    count_accounts,
)

ACCOUNT_DB_PATH = PROJECT_ROOT / "account_data" / "cache" / "account.sqlite3"
STATIC_DB_PATH = PROJECT_ROOT / "data" / "genshin.sqlite3"
init_db(STATIC_DB_PATH)
init_account_db(ACCOUNT_DB_PATH)
Session = make_account_session_factory(ACCOUNT_DB_PATH)
StaticSession = make_session_factory(STATIC_DB_PATH)

ACCOUNT_DB_PATH
```

## 4. 保存到账号库

同一个 UID 下同名角色会被更新。武器和圣遗物会作为账号背包物品写入，并挂到这个角色身上。

```python
SLOT_TO_EQUIPPED_FIELD = {
    "flower": "equipped_flower_id",
    "plume": "equipped_plume_id",
    "sands": "equipped_sands_id",
    "goblet": "equipped_goblet_id",
    "circlet": "equipped_circlet_id",
}


def require_text(value: str, field_name: str) -> str:
    value = str(value or "").strip()
    if not value:
        raise ValueError(f"请先填写 {field_name}")
    return value


def find_artifact_set_name_by_piece(piece_name: str) -> str | None:
    piece_name = str(piece_name or "").strip()
    if not piece_name:
        return None

    with StaticSession() as static_session:
        artifact_set = static_session.scalar(
            select(ArtifactSet).where(
                or_(
                    ArtifactSet.flower_name == piece_name,
                    ArtifactSet.plume_name == piece_name,
                    ArtifactSet.sands_name == piece_name,
                    ArtifactSet.goblet_name == piece_name,
                    ArtifactSet.circlet_name == piece_name,
                )
            )
        )
    if artifact_set is None:
        return None
    return artifact_set.name


uid = require_text(ACCOUNT_UID, "ACCOUNT_UID")
character_name = require_text(CHARACTER.get("character_name"), "CHARACTER['character_name']")

with Session.begin() as session:
    account = session.get(Account, uid)
    if account is None:
        account = Account(uid=uid)
        session.add(account)

    character = session.scalar(
        select(AccountCharacter).where(
            AccountCharacter.uid == uid,
            AccountCharacter.character_name == character_name,
        )
    )
    if character is None:
        character = AccountCharacter(uid=uid, character_name=character_name)
        session.add(character)

    for field in [
        "level",
        "ascension",
        "constellation",
        "normal_attack_level",
        "elemental_skill_level",
        "elemental_burst_level",
    ]:
        setattr(character, field, CHARACTER.get(field))

    weapon_name = str(EQUIPPED_WEAPON.get("weapon_name") or "").strip()
    if weapon_name:
        weapon = AccountWeapon(
            uid=uid,
            weapon_name=weapon_name,
            level=EQUIPPED_WEAPON.get("level"),
            refinement_rank=EQUIPPED_WEAPON.get("refinement_rank"),
        )
        session.add(weapon)
        session.flush()
        character.equipped_weapon_id = weapon.id

    for artifact_data in EQUIPPED_ARTIFACTS:
        slot = require_text(artifact_data.get("slot"), "artifact['slot']")
        if slot not in SLOT_TO_EQUIPPED_FIELD:
            valid_slots = ", ".join(SLOT_TO_EQUIPPED_FIELD)
            raise ValueError(f"未知圣遗物部位：{slot}。可用值：{valid_slots}")

        piece_name = artifact_data.get("piece_name") or None
        set_name = find_artifact_set_name_by_piece(piece_name) if piece_name else None
        if piece_name and set_name is None:
            raise ValueError(
                f"未能从静态圣遗物库找到单件名称：{piece_name}。"
                "请检查名称是否准确，或先运行 genshin-analyser sync-artifacts。"
            )

        artifact = AccountArtifact(
            uid=uid,
            set_name=set_name,
            piece_name=piece_name,
            slot=slot,
            level=artifact_data.get("level"),
            rarity=artifact_data.get("rarity"),
            main_stat=artifact_data.get("main_stat") or {},
            substats=artifact_data.get("substats") or [],
        )
        session.add(artifact)
        session.flush()
        setattr(character, SLOT_TO_EQUIPPED_FIELD[slot], artifact.id)

print(f"已保存账号 {uid} 的角色：{character_name}")
```

## 5. 查看录入结果

```python
with Session() as session:
    print(f"账号数量：{count_accounts(session)}")
    print(f"账号角色数量：{count_account_characters(session)}")
    print(f"账号武器数量：{count_account_weapons(session)}")
    print(f"账号圣遗物数量：{count_account_artifacts(session)}")

    rows = session.scalars(
        select(AccountCharacter)
        .where(AccountCharacter.uid == ACCOUNT_UID)
        .order_by(AccountCharacter.character_name)
    ).all()
    for row in rows:
        print(
            row.character_name,
            f"等级 {row.level or '-'}",
            f"命座 {row.constellation if row.constellation is not None else '-'}",
            f"天赋 {row.normal_attack_level or '-'}/{row.elemental_skill_level or '-'}/{row.elemental_burst_level or '-'}",
        )
```
