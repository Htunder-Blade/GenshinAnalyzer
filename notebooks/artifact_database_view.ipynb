# 圣遗物数据库验证

这个 Notebook 用来验证本地原神圣遗物数据库。当前圣遗物表只关注套装名称、各个部位的具体名字和套装效果。

```python
from __future__ import annotations

import sqlite3
from pathlib import Path

from IPython.display import Markdown, display


def find_project_root(start: Path) -> Path:
    for candidate in [start, *start.parents]:
        if (candidate / "data").exists():
            return candidate
    return start


project_root = find_project_root(Path.cwd())
db_candidates = [
    project_root / "data" / "genshin.sqlite3",
    project_root / "data" / "genshin_characters.sqlite3",
]
db_path = next((candidate for candidate in db_candidates if candidate.exists()), db_candidates[0])
artifact_cache_dir = project_root / "data" / "cache" / "artifacts"


def connect():
    connection = sqlite3.connect(db_path)
    connection.row_factory = sqlite3.Row
    return connection


def fetch_all(sql, params=()):
    with connect() as connection:
        return [dict(row) for row in connection.execute(sql, params)]


def fetch_one(sql, params=()):
    with connect() as connection:
        row = connection.execute(sql, params).fetchone()
        return dict(row) if row is not None else None


def markdown_table(headers, rows):
    if not rows:
        return '\u65e0\u6570\u636e'
    header = "| " + " | ".join(headers) + " |"
    divider = "| " + " | ".join(["---"] * len(headers)) + " |"
    body = ["| " + " | ".join(str(item if item is not None else "-").replace("|", "\\|") for item in row) + " |" for row in rows]
    return "\n".join([header, divider, *body])


print(f"\u6570\u636e\u5e93\u8def\u5f84\uff1a{db_path}")
print(f"\u6570\u636e\u5e93\u5b58\u5728\uff1a{db_path.exists()}")
print(f"\u5723\u9057\u7269\u7f13\u5b58\u76ee\u5f55\uff1a{artifact_cache_dir}")
print(f"\u5723\u9057\u7269\u7f13\u5b58\u76ee\u5f55\u5b58\u5728\uff1a{artifact_cache_dir.exists()}")
if not db_path.exists():
    raise FileNotFoundError(f"\u6ca1\u6709\u627e\u5230\u6570\u636e\u5e93\uff1a{db_path}")
```

## 1. 总体验证

```python
expected_columns = [
    "name",
    "flower_name",
    "plume_name",
    "sands_name",
    "goblet_name",
    "circlet_name",
    "effect_1pc",
    "effect_2pc",
    "effect_4pc",
]
expected_count = 61

table_exists = fetch_one("select count(*) as count from sqlite_master where type = 'table' and name = 'artifact_sets'")["count"]
columns = fetch_all("pragma table_info(artifact_sets)") if table_exists else []
actual_columns = [row["name"] for row in columns]
primary_key_columns = [row["name"] for row in columns if row["pk"]]
summary = fetch_one("select count(*) as total, count(distinct name) as unique_names from artifact_sets")
duplicate_groups = fetch_one("""
select count(*) as count
from (select name from artifact_sets group by name having count(*) > 1)
""")["count"]

checks = [
    ['artifact_sets \u8868\u5b58\u5728', table_exists == 1, table_exists],
    ['\u5b57\u6bb5\u53ea\u5305\u542b\u5957\u88c5\u540d\u3001\u90e8\u4f4d\u540d\u3001\u5957\u88c5\u6548\u679c', actual_columns == expected_columns, ", ".join(actual_columns)],
    ['name \u662f\u4e3b\u952e', primary_key_columns == ["name"], ", ".join(primary_key_columns)],
    ['\u5957\u88c5\u6570\u91cf\u4e3a 61', summary["total"] == expected_count, summary["total"]],
    ['\u5957\u88c5\u540d\u79f0\u65e0\u91cd\u540d', duplicate_groups == 0, duplicate_groups],
]

display(Markdown(markdown_table(['\u68c0\u67e5\u9879', '\u662f\u5426\u901a\u8fc7', '\u5f53\u524d\u503c'], checks)))
```

## 2. 部位和效果完整性

```python
normal_missing_pieces = fetch_one("""
select count(*) as count
from artifact_sets
where effect_1pc is null and (
    flower_name is null or plume_name is null or sands_name is null or goblet_name is null or circlet_name is null
)
""")["count"]
one_piece_count = fetch_one("select count(*) as count from artifact_sets where effect_1pc is not null")["count"]
one_piece_bad_rows = fetch_one("""
select count(*) as count
from artifact_sets
where effect_1pc is not null and (
    circlet_name is null or flower_name is not null or plume_name is not null or sands_name is not null or goblet_name is not null
)
""")["count"]
missing_effects = fetch_one("""
select count(*) as count
from artifact_sets
where effect_1pc is null and (effect_2pc is null or effect_4pc is null)
""")["count"]

checks = [
    ['\u5e38\u89c4\u5957\u88c5\u4e94\u4e2a\u90e8\u4f4d\u5747\u5b8c\u6574', normal_missing_pieces == 0, normal_missing_pieces],
    ['\u5355\u4ef6\u5957\u6570\u91cf\u4e3a 4', one_piece_count == 4, one_piece_count],
    ['\u5355\u4ef6\u5957\u53ea\u4fdd\u7559\u7406\u4e4b\u51a0\u548c 1 \u4ef6\u6548\u679c', one_piece_bad_rows == 0, one_piece_bad_rows],
    ['\u5e38\u89c4\u5957\u88c5 2/4 \u4ef6\u6548\u679c\u5747\u5b8c\u6574', missing_effects == 0, missing_effects],
]

one_piece_rows = fetch_all("""
select name, circlet_name, effect_1pc
from artifact_sets
where effect_1pc is not null
order by name
""")

content = [
    markdown_table(['\u68c0\u67e5\u9879', '\u662f\u5426\u901a\u8fc7', '\u5f53\u524d\u503c'], checks),
    "### \u5355\u4ef6\u5957",
    markdown_table(['\u5957\u88c5', '\u7406\u4e4b\u51a0', '1 \u4ef6\u6548\u679c'], [[row['name'], row['circlet_name'], row['effect_1pc']] for row in one_piece_rows]),
]
display(Markdown("\n\n".join(content)))
```

## 3. 套装列表与检索

```python
keyword = ""
limit = 50

where = "where name like ?" if keyword else ""
params = [f"%{keyword}%"] if keyword else []
sql = f"""
select name, effect_1pc, effect_2pc, effect_4pc
from artifact_sets
{where}
order by name
limit ?
"""
params.append(limit)
rows = fetch_all(sql, params)

display(Markdown(markdown_table(
    ['\u5957\u88c5', '1 \u4ef6\u6548\u679c', '2 \u4ef6\u6548\u679c', '4 \u4ef6\u6548\u679c'],
    [[row['name'], row['effect_1pc'], row['effect_2pc'], row['effect_4pc']] for row in rows],
)))
```

## 4. 单套详情

```python
artifact_set_name = "逐影猎人"

artifact_set = fetch_one("select * from artifact_sets where name = ?", (artifact_set_name,)) if artifact_set_name else None
if artifact_set is None:
    artifact_set = fetch_one("select * from artifact_sets order by name limit 1")
if artifact_set is None:
    raise ValueError("\u6570\u636e\u5e93\u4e2d\u6ca1\u6709\u5723\u9057\u7269\u5957\u88c5\u6570\u636e")

piece_rows = [
    ['\u751f\u4e4b\u82b1', artifact_set['flower_name']],
    ['\u6b7b\u4e4b\u7fbd', artifact_set['plume_name']],
    ['\u65f6\u4e4b\u6c99', artifact_set['sands_name']],
    ['\u7a7a\u4e4b\u676f', artifact_set['goblet_name']],
    ['\u7406\u4e4b\u51a0', artifact_set['circlet_name']],
]
effect_rows = [
    ['1 \u4ef6', artifact_set['effect_1pc']],
    ['2 \u4ef6', artifact_set['effect_2pc']],
    ['4 \u4ef6', artifact_set['effect_4pc']],
]

content = [
    f"### {artifact_set['name']}",
    "#### \u90e8\u4f4d\u540d\u79f0",
    markdown_table(['\u90e8\u4f4d', '\u540d\u79f0'], piece_rows),
    "#### \u5957\u88c5\u6548\u679c",
    markdown_table(['\u4ef6\u6570', '\u6548\u679c'], effect_rows),
]
display(Markdown("\n\n".join(content)))
```
