# 角色数据库查看

这个 Notebook 用来查看本地原神角色数据库的统计信息，并按名称、元素、武器、地区等条件检索角色。

这版本恢复使用 display(Markdown(...)) 来展示表格。如果 VS Code 里 Run All 仍然出现 `notebook controller is DISPOSED`，优先按 VS Code Jupyter 的 kernel/controller 状态问题处理。


```python
from __future__ import annotations

import json
import sqlite3
from pathlib import Path

from IPython.display import Markdown, display

# 统一定位项目根目录，从项目根目录或 notebooks 目录打开都能找到数据库。
project_root = Path.cwd() if (Path.cwd() / "data").exists() else Path.cwd().parent
db_path = project_root / "data" / "genshin.sqlite3"


def connect():
    connection = sqlite3.connect(db_path)
    connection.row_factory = sqlite3.Row
    return connection


def fetch_all(sql, params=()):
    with connect() as connection:
        return [dict(row) for row in connection.execute(sql, params)]


def fetch_one(sql, params=()):
    with connect() as connection:
        row = connection.execute(sql, params).fetchone()
        return dict(row) if row is not None else None


def markdown_table(headers, rows):
    # 使用 Markdown 表格输出，这是上一版 DataFrame 之前的展示方式。
    if not rows:
        return '\u65e0\u6570\u636e'
    header = "| " + " | ".join(headers) + " |"
    divider = "| " + " | ".join(["---"] * len(headers)) + " |"
    body = ["| " + " | ".join(str(item if item is not None else "-").replace("|", "\|") for item in row) + " |" for row in rows]
    return "\n".join([header, divider, *body])

print(f"数据库路径?{db_path}")
print(f"数据库存在?{db_path.exists()}")
if not db_path.exists():
    raise FileNotFoundError(f"没有找到数据库?{db_path}")
```

## 1. 总览统计


```python
# 这里检查角色总数，以及旅行者空/荧是否已经按当前阶段要求排除。
summary = fetch_one("""
select
  count(*) as total_characters,
  sum(case when name in ('\u7a7a', '\u8367') then 1 else 0 end) as traveler_remaining
from characters
""")

display(Markdown(markdown_table(
    ['\u9879\u76ee', '\u6570\u91cf'],
    [
        ['\u89d2\u8272\u603b\u6570', summary["total_characters"]],
        ['\u65c5\u884c\u8005\u7a7a/\u8367\u6b8b\u7559\u6570\u91cf', summary["traveler_remaining"]],
    ],
)))
```

## 2. 分类统计


```python
# Aggregate by common slim character fields. NULL values are displayed as "-".
def stat_by(field: str):
    rows = fetch_all(f"""
    select coalesce({field}, '-') as category, count(*) as count
    from characters
    group by coalesce({field}, '-')
    order by count desc, category
    """)
    return [[row["category"], row["count"]] for row in rows]

sections = [
    ('按元素统计', "element"),
    ('按武器统计', "weapon_type"),
    ('按星级统计', "rarity"),
]

content = []
for title, field in sections:
    content.append(f"### {title}")
    content.append(markdown_table(['分类', '数量'], stat_by(field)))

display(Markdown("\n\n".join(content)))
```

## 3. 角色图鉴列表

可以修改下面几个筛选条件。留空表示不过滤。


```python
# Change these variables to filter the character list.
keyword = ""
element = ""
weapon = ""
rarity = ""
limit = 50

conditions = []
params = []
if keyword:
    conditions.append("name like ?")
    params.append(f"%{keyword}%")
if element:
    conditions.append("element = ?")
    params.append(element)
if weapon:
    conditions.append("weapon_type = ?")
    params.append(weapon)
if rarity:
    conditions.append("rarity = ?")
    params.append(str(rarity))

where = " where " + " and ".join(conditions) if conditions else ""
sql = f"""
select name, rarity, element, weapon_type
from characters
{where}
order by element, name
limit ?
"""
params.append(limit)
rows = fetch_all(sql, params)

table_rows = [
    [row["name"], row["rarity"], row["element"], row["weapon_type"]]
    for row in rows
]
display(Markdown(markdown_table(['名称', '星级', '元素', '武器'], table_rows)))
```

## 4. 查看单个角色详情


```python
# Set character_name to inspect a specific character. If it is missing, show the first row.
character_name = '钟离'

if character_name:
    character = fetch_one("select * from characters where name = ?", (character_name,))
else:
    character = None
if character is None:
    character = fetch_one("select * from characters order by element, name limit 1")
if character is None:
    raise ValueError("数据库中没有角色数据")

detail_rows = [
    ['名称', character["name"]],
    ['星级', character["rarity"]],
    ['元素', character["element"]],
    ['武器', character["weapon_type"]],
]
display(Markdown(markdown_table(['字段', '内容'], detail_rows)))
```

## 5. 天赋、命座与关键等级面板


```python
# 角色的天赋、命座、等级面板保存在 JSON 字段里，这里只抽取最常用的摘要信息。
import re

character_data = json.loads(character["character_data"])
talent_data = json.loads(character["talent_data"])
constellation_data = json.loads(character["constellation_data"])
stats_data = json.loads(character["stats_data"])


def one_line(value, max_len=80):
    text = str(value or "-").replace("\n", " ").replace("|", "\|")
    return text if len(text) <= max_len else text[:max_len] + "..."


def format_specialized(value, substat_type):
    if value in (None, ""):
        return "-"
    if substat_type == "FIGHT_PROP_ELEMENT_MASTERY":
        return f"{float(value):.0f}"
    return f"{float(value) * 100:.2f}%"


def format_talent_value(value, spec):
    if value in (None, ""):
        return "-"
    number = float(value)
    if spec.endswith("P"):
        decimals = _format_decimals(spec, default=1)
        return f"{number * 100:.{decimals}f}%"
    if spec.endswith("I"):
        return str(int(round(number)))
    decimals = _format_decimals(spec, default=1)
    return f"{number:.{decimals}f}"


def _format_decimals(spec, default=1):
    match = re.search(r"F(\d+)", spec)
    return int(match.group(1)) if match else default


def render_talent_attribute(template, params, level):
    index = level - 1

    def replace(match):
        key = match.group("key")
        spec = match.group("spec") or "F1"
        values = params.get(key, [])
        value = values[index] if index < len(values) else None
        return format_talent_value(value, spec)

    return re.sub(r"\{(?P<key>param\d+)(?::(?P<spec>[^}]+))?\}", replace, template)

result_talents = talent_data.get("result", {})
talent_rows = []
talent_value_rows = []
talent_levels = [1, 6, 9, 10, 13]
for label, key in [
    ('\u666e\u901a\u653b\u51fb', "combat1"),
    ('\u5143\u7d20\u6218\u6280', "combat2"),
    ('\u5143\u7d20\u7206\u53d1', "combat3"),
    ('\u56fa\u6709\u5929\u8d4b 1', "passive1"),
    ('\u56fa\u6709\u5929\u8d4b 2', "passive2"),
    ('\u56fa\u6709\u5929\u8d4b 3', "passive3"),
]:
    item = result_talents.get(key, {})
    if item:
        talent_rows.append([label, item.get("name", "-"), one_line(item.get("description"))])
        attributes = item.get("attributes") or {}
        params = attributes.get("parameters") or {}
        for raw_label in attributes.get("labels") or []:
            attr_name, template = raw_label.split("|", 1) if "|" in raw_label else (raw_label, raw_label)
            talent_value_rows.append([
                label,
                item.get("name", "-"),
                attr_name,
                *[render_talent_attribute(template, params, level) for level in talent_levels],
            ])

result_constellations = constellation_data.get("result", {})
constellation_rows = []
for index in range(1, 7):
    item = result_constellations.get(f"c{index}", {})
    if item:
        constellation_rows.append([f"C{index}", item.get("name", "-"), one_line(item.get("description"))])

character_result = character_data.get("result", {})
substat_name = character_result.get("substatText") or "-"
substat_type = character_result.get("substatType")

preferred_levels = ["1", "20", "40", "50", "60", "70", "80", "90"]
stat_rows = []
for level in preferred_levels:
    item = stats_data.get(level)
    if item:
        stat_rows.append([
            level,
            item.get("ascension", "-"),
            round(item.get("hp", 0), 1),
            round(item.get("attack", 0), 1),
            round(item.get("defense", 0), 1),
            f"{substat_name} {format_specialized(item.get('specialized'), substat_type)}",
        ])

content = [
    "### " + '\u5929\u8d4b\u6458\u8981',
    markdown_table(['\u7c7b\u578b', '\u540d\u79f0', '\u8bf4\u660e'], talent_rows),
    "### " + '\u5929\u8d4b\u7b49\u7ea7\u6570\u503c',
    markdown_table(['\u7c7b\u578b', '\u540d\u79f0', '\u6761\u76ee', *[f"Lv.{level}" for level in talent_levels]], talent_value_rows),
    "### " + '\u547d\u5ea7\u6458\u8981',
    markdown_table(['\u547d\u5ea7', '\u540d\u79f0', '\u8bf4\u660e'], constellation_rows),
    "### " + '\u5173\u952e\u7b49\u7ea7\u9762\u677f',
    markdown_table(['\u7b49\u7ea7', '\u7a81\u7834', '\u751f\u547d\u503c', '\u653b\u51fb\u529b', '\u9632\u5fa1\u529b', '\u7a81\u7834\u5c5e\u6027'], stat_rows),
]
display(Markdown("\n\n".join(content)))
```

## 7. 武器数据库总览

```python
# Weapon database overview. Weapon names are unique after excluding the three duplicated Isshin blades.
weapon_count = fetch_one("select count(*) as count from weapons")["count"]
weapon_type_rows = fetch_all("""
select coalesce(weapon_type, '-') as category, count(*) as count
from weapons
group by coalesce(weapon_type, '-')
order by count desc, category
""")
weapon_rarity_rows = fetch_all("""
select coalesce(rarity, '-') as category, count(*) as count
from weapons
group by coalesce(rarity, '-')
order by cast(category as integer) desc
""")
duplicate_name_count = fetch_one("""
select count(*) as count
from (
    select name
    from weapons
    group by name
    having count(*) > 1
)
""")["count"]

content = [
    f"武器数量：{weapon_count}",
    f"重名武器组数：{duplicate_name_count}",
    "### " + '按武器类型统计',
    markdown_table(['类型', '数量'], [[row["category"], row["count"]] for row in weapon_type_rows]),
    "### " + '按星级统计',
    markdown_table(['星级', '数量'], [[row["category"], row["count"]] for row in weapon_rarity_rows]),
]
display(Markdown("\n\n".join(content)))
```

## 8. 武器列表与检索

```python
# Change these variables to filter the weapon list.
weapon_keyword = ""
weapon_type = ""
weapon_rarity = ""
weapon_limit = 50

conditions = []
params = []
if weapon_keyword:
    conditions.append("name like ?")
    params.append(f"%{weapon_keyword}%")
if weapon_type:
    conditions.append("weapon_type = ?")
    params.append(weapon_type)
if weapon_rarity:
    conditions.append("rarity = ?")
    params.append(str(weapon_rarity))

where = " where " + " and ".join(conditions) if conditions else ""
sql = f"""
select name, rarity, weapon_type, weapon_data
from weapons
{where}
order by weapon_type, cast(rarity as integer) desc, name
limit ?
"""
params.append(weapon_limit)
weapon_rows = fetch_all(sql, params)

table_rows = []
for row in weapon_rows:
    data = json.loads(row["weapon_data"]).get("result", {})
    table_rows.append([
        row["name"],
        row["rarity"],
        row["weapon_type"],
        data.get("mainStatText") or "-",
        data.get("effectName") or "-",
    ])

display(Markdown(markdown_table(['名称', '星级', '类型', '副属性', '特效'], table_rows)))
```

## 9. 查看单把武器

```python
# Set weapon_name to inspect a specific weapon. If missing, show the first weapon.
weapon_name = "护摩之杖"

if weapon_name:
    weapon = fetch_one("select * from weapons where name = ?", (weapon_name,))
else:
    weapon = None
if weapon is None:
    weapon = fetch_one("select * from weapons order by weapon_type, cast(rarity as integer) desc, name limit 1")
if weapon is None:
    raise ValueError("数据库中没有武器数据")

weapon_data = json.loads(weapon["weapon_data"])
weapon_stats_data = json.loads(weapon["stats_data"])
weapon_result = weapon_data.get("result", {})

def format_weapon_specialized(value, stat_type):
    if value in (None, ""):
        return "-"
    if stat_type == "FIGHT_PROP_ELEMENT_MASTERY":
        return f"{float(value):.0f}"
    return f"{float(value) * 100:.2f}%"

weapon_detail_rows = [
    ['名称', weapon["name"]],
    ['星级', weapon["rarity"]],
    ['类型', weapon["weapon_type"]],
    ['副属性', weapon_result.get("mainStatText") or "-"],
    ['特效', weapon_result.get("effectName") or "-"],
]

weapon_preferred_levels = ["1", "20", "40", "50", "60", "70", "80", "90"]
weapon_stat_rows = []
main_stat_name = weapon_result.get("mainStatText") or "-"
main_stat_type = weapon_result.get("mainStatType")
for level in weapon_preferred_levels:
    item = weapon_stats_data.get(level)
    if item:
        weapon_stat_rows.append([
            level,
            item.get("ascension", "-"),
            round(item.get("attack", 0), 1),
            f"{main_stat_name} {format_weapon_specialized(item.get('specialized'), main_stat_type)}",
        ])

refinement_rows = []
for refinement in range(1, 6):
    item = weapon_result.get(f"r{refinement}") or {}
    refinement_rows.append([
        f"R{refinement}",
        one_line(item.get("description"), max_len=120),
        ", ".join(item.get("values") or []) or "-",
    ])

content = [
    "### " + '武器基本信息',
    markdown_table(['字段', '内容'], weapon_detail_rows),
    "### " + '关键等级面板',
    markdown_table(['等级', '突破', '基础攻击力', '副属性'], weapon_stat_rows),
    "### " + '精炼效果',
    markdown_table(['精炼', '说明', '数值'], refinement_rows),
]
display(Markdown("\n\n".join(content)))
```
